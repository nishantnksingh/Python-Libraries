# Pandas_Indexing_and_Selection

 Indexing in Series (Indexers: loc, iloc)

* Selecting rows: ```df[start:stop]```
* Selecting columns: ```df['column']``` or ```df.column``` or ```df[['col_x', 'col_y']]```
    * ```df['column']``` or ```df.column``` return a series
    * ```df[['col_x', 'col_y']]``` returns a dataframe

### A. loc()    
Uses Location of the value for indexing

````python
import pandas as pd
value = pd.Series(['a', 'b', 'c', 'e', 'f'], index=[1, 2, 3, 4, 5])
value
````

````python
value.loc[1]
````

````python
value.loc[1:3]
````

````python
print(info)
````

````python
info.loc['New York','population']
````

 # sorting Dataframes 

sorting can be done in two ways
1. Sorting by indicies 
2. sorting by values 

````python
import pandas as pd 
````

````python
df = pd.read_csv("C:\\Users\\Hp\\Desktop\\Nikhat\\datasets\\Housing.csv")
````

````python
df
````

````python
df.head()
````

````python
# setting price as index 
df.set_index("stories",inplace = True)
````

````python
df.head()
````

#### sorting by index 

````python
df.sort_index (axis = 0, ascending  = False ).head(15)
````

#### sorting by values 

````python
# instead of making price as index we can directly sort the values of price by this method 
df.reset_index(inplace = True )
````

````python
df.head()
````

````python
# sorting by price 
df.sort_values(by = "price", ascending = True ).head(10)
````

````python
# sorting by area 
df.sort_values(by = "area", ascending = False ).head(10)
````

````python
# sorting multiple columns 

df.sort_values(by = ["bedrooms", "price"], ascending = False ) 
````

````python
df.sort_values(by = ["price", "bedrooms"], ascending = True)
````

Filetering rows of the dataframe based on conditions like 

- price < 5000000
- bedrooms > 3

now we can use .loc for extracrion of rows 

````python
df.loc[df.price < 5000000]
````

````python
df.loc[df["price"]< 5000000]
````

#### filetring based on multiple conditions using "&" ,"or(|)" operaters

````python

df.loc[(df.price > 4000000) & (df.bedrooms==3)]
````

````python
df.loc[(df.bedrooms>3) | (df.stories<1)]
````

````python
# filtering rows on condition and selecting only required columns 
df.loc[(df.bedrooms>2) & (df.stories == 1 ), ["price", "area","parking"]]
````

````python
df.loc[(df.bedrooms == 4)]
````

## 02_Matplotlib_Multiple Plots_Subplots_Gridplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### 2.1 Multiple Plots

```python
data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20, 'pineapple': 34,
       'banana': 24, 'grapes': 65}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(15,4), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

### 2.2 Multiple Plots

```python
X = np.linspace(0, 10, 1000)
Y1 = np.sin(X)
Y2 = np.cos(X)


fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharey=True)
fig.suptitle('Multiple Plotting')

axs[0].plot(X, Y1, c='g')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Sin(X)')

axs[1].plot(X, Y2, c='y')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Cos(X)')

plt.show()
```

### 2.3 Multiple Plots

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)


fig, axs = plt.subplots(2, 2, figsize=(15, 8), sharey=True)

axs[0,0].plot(X, Y, '--g')
axs[0,0].set_title('Axis [0,0]')
axs[0,0].set_xlabel('X1')
axs[0,0].set_ylabel('Y1)')

axs[0,1].plot(X, Y, 'y^', )
axs[0,1].set_title('Axis [0,1]')
axs[0,1].set_xlabel('X2')
axs[0,1].set_ylabel('Y2)')

axs[1,0].plot(X, Y, '.b')
axs[1,0].set_title('Axis [1,0]')
axs[1,0].set_xlabel('X3')
axs[1,0].set_ylabel('Y3)')

axs[1,1].plot(X, Y, '+-r')
axs[1,1].set_title('Axis [1,1]')
axs[1,1].set_xlabel('X4')
axs[1,1].set_ylabel('Y4')

fig.suptitle('Multiple Plotting')
```

### 1.4 Grid Plots

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

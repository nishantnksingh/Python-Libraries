## 01_Matplotlib_Plotting

```python
%matplotlib inline
import numpy as np
```

### 1. Line Plots

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot([1, 2, 4, 7, 9], [1, 6, 4, 3, 8])    # line graph
plt.show()
```

```python
plt.axis([0, 7, 0, 7])   # axis is 0-7

plt.plot([1, 2, 4, 7, 9], [1, 6, 4, 3, 8])
plt.show()
```

```python
X = np.linspace(-2, 2, 500)
y = X**2

plt.plot(X, y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

### A. Adjusting the Plot: Line Colors and Styles

```python
x = np.linspace(0, 10, 100)

plt.axis([0, 10, -1, 1])

plt.plot(x, np.sin(x - 0), c = 'b')              # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 1), c = 'green')          # specify color by name
plt.plot(x, np.sin(x - 2), c = '0.75')           # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')      # Hex code (RRGGBB from 00 to FF)

plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.axis([0, 10, -1, 1])
 
plt.plot(x, np.sin(x - 0), '-g')         # solid green            
plt.plot(x, np.sin(x - 1), '--c')        # dashed cyan
plt.plot(x, np.sin(x - 2), '-.k')        # dashdot black  
plt.plot(x, np.sin(x - 3), ':r')         # dotted red

plt.show()
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, 2*x + 2, '-.k') # dashdot black
plt.plot(x, x**2 + 3, ':r');  # dotted red
```

### B. Plot Limits

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red

plt.xlim(0, 10)
plt.ylim(0, 12);
```

### C. Labeling Plots and Grid

```python
plt.plot(x, np.sin(x))

plt.title("Sine function")
plt.xlabel("x", fontsize = 15)
plt.ylabel("sine(x)", fontsize = 15)

plt.grid(True)
plt.show()
```

### D. Plot Legends

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':r', label='cos(x)')
plt.axis([0, 10, -3, 3])

plt.title("Sine function")
plt.xlabel("x", fontsize = 15)
plt.ylabel("y", fontsize = 15)

plt.grid(True)
plt.legend(loc='best');
```

### E. Figure size

```python
plt.figure(figsize=(10,6))

plt.plot(x, np.sin(x), '-g', label='sin(x)', marker = 'o')
plt.plot(x, np.cos(x), ':r', label='cos(x)')
plt.axis([0, 10, -3, 3])

plt.grid(True)
plt.legend(loc='best');
```

## 2. Scatter Plots

```python
A = np.linspace(0, 10, 50) 
B = np.cos(A)
C = np.sin(A)

plt.scatter(A, B, marker = 'o', color = 'r');
plt.scatter(A, C, marker = '^', color = 'g');
```

```python
np.random.seed(0)

xs = np.random.randn(500)
ys = np.random.randn(500)

plt.scatter(xs, ys);
```

###  A. ``alpha`` key

```python
plt.scatter(xs, ys, alpha = 0.5);
```

###  B. ``s`` key

```python
scale = 200 * (np.random.rand(100)) 

plt.scatter(xs, ys, s = scale, alpha = 0.3);
plt.grid(True)
```

###  C. ``c`` and ``cmap`` key

```python
scale = 200 * (np.random.rand(100)) 
color = np.random.rand(500)

plt.scatter(xs, ys, s = scale, c=color, alpha = 0.3, cmap = 'ocean');
plt.grid(True)
plt.colorbar()
```

## 3. Histograms

### A. One-Dimensional Histograms 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000); 
plt.hist(data);
```

```python
data = np.random.randn(1000); 

plt.hist(data, bins=50, alpha=0.5, histtype='step', color='r');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40, edgecolor='k')

plt.hist(x1, **kwargs, label = 'one')
plt.hist(x2, **kwargs, label = 'two')
plt.hist(x3, **kwargs, label = 'three')

plt.legend();
```

### B. Two-Dimensional Histograms

```python
x1 = np.random.normal(0, 0.8, 10000)
x2 = np.random.normal(-2, 1, 10000)

plt.hist2d(x1, x2, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x1, x2, gridsize=60, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

## 4. Bar Chart

```python
data = {'A': 10, 'B': 15, 'C': 5, 'D': 20, 'E': 34,
       'F': 24, 'G': 65, 'H': 45, 'I': 23}
names = list(data.keys())
values = list(data.values())


```

```python
plt.bar(names, values, color='k')
plt.xlabel('Names')
plt.ylabel('Values')
plt.show();
```

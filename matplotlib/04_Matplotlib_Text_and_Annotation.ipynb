# 04_Matplotlib_Text_and_Annotation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
```

## 1. Adding Text

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x))

plt.text(4.05, -0.75, "Low value", size=15, color='k')
plt.text(7.22, 0.6, "High value", size=15, color='g')

plt.title("Sine function", fontsize = 20)
plt.xlabel("x", fontsize = 15)
plt.ylabel("sine(x)", fontsize = 15)

plt.show()
```

## 2. Arrows and Annotation

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x))

plt.annotate('local minimum', xy=(4.6, -1), xytext=(6, -0.5),
            arrowprops=dict(arrowstyle="->"))
plt.annotate('local maximum', xy=(7.8, 1), xytext=(7, 0.5),
            arrowprops=dict(facecolor='crimson', shrink=1))


plt.title("Sine function", fontsize = 20)
plt.xlabel("x", fontsize = 15)
plt.ylabel("sine(x)", fontsize = 15)

plt.show()
```

## 3. Lines

```python
np.random.seed(0)

xs = np.random.randn(500)
ys = np.random.randn(500)

plt.figure(figsize=(10,6))
plt.scatter(xs, ys);

plt.plot(1, -1, "y^")
plt.vlines(1, -3, -1, color="green")
plt.hlines(-1, -5, 1, color="red", linestyles='--')

plt.axis([-3, 3, -3, 3])

plt.show()
```
